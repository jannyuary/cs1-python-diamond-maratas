```python
# Input student's name and score
name = input("Enter student's name: ")
score = int(input("Input: "))

# Determine the letter grade
if 90 <= score <= 100:
    grade = "A"
elif 80 <= score <= 89:
    grade = "B"
elif 70 <= score <= 79:
    grade = "C"
else:
    grade = "F"

# Display the result
print("Output:", grade)
```
