```python
# Student Grade Manager
# Demonstrates: Input/Output, Control Structures, Logic, Functions

def get_grade(score):
    """Function to determine letter grade"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"


def get_average(scores):
    """Function to compute average"""
    return sum(scores) / len(scores)


def display_result(name, scores):
    """Function to display results"""
    avg = get_average(scores)
    grade = get_grade(avg)

    print("\n--- Student Result ---")
    print("Name:", name)
    print("Scores:", scores)
    print("Average:", round(avg, 2))
    print("Final Grade:", grade)


def main():
    print("=== Student Grade Manager ===")

    name = input("Enter student name: ")

    scores = []
    num_subjects = int(input("How many subjects? "))

    # Loop for input
    for i in range(num_subjects):
        while True:
            try:
                score = float(input(f"Enter score for subject {i+1}: "))

                # Logic validation
                if 0 <= score <= 100:
                    scores.append(score)
                    break
                else:
                    print("Score must be between 0 and 100.")
            except:
                print("Invalid input. Please enter a number.")

    display_result(name, scores)

    # Control structure (decision)
    choice = input("\nDo you want to enter another student? (yes/no): ").lower()

    if choice == "yes":
        main()
    else:
        print("Program ended. Thank you!")


# Run program
main()
```
